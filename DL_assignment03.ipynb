```python
from ultralytics import YOLO

# Load YOLOv8 nano model
model = YOLO('yolov8n.pt')

# Train model
model.train(
    data='coco128.yaml',
    epochs=10,
    imgsz=640,
    batch=16,
    name='yolov8_detection'
)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict(source='https://ultralytics.com/images/bus.jpg', save=True)
```

```python
from ultralytics import YOLO

seg_model = YOLO('yolov8n-seg.pt')

seg_model.train(
    data='coco128-seg.yaml',
    epochs=10,
    imgsz=640,
    name='yolov8_segmentation'
)
```

```python
seg_model.predict(
    source='https://ultralytics.com/images/bus.jpg',
    save=True
)
```

```python
from ultralytics import YOLO

cls_model = YOLO('yolov8n-cls.pt')

cls_model.train(
    data='mnist160',
    epochs=5,
    imgsz=64,
    name='yolov8_classification'
)
```

```python
cls_model.predict(
    source='https://ultralytics.com/images/bus.jpg',
    save=True
)
```

Q1 Answer

Detection predicts bounding boxes, segmentation predicts pixel-level masks, and classification predicts only class labels. Detection and segmentation use similar backbone networks but segmentation includes a mask prediction head. Classification only uses feature extraction and classification layers.

Q2 Answer

Changing learning rate significantly affected training stability. A high learning rate caused unstable training and lower accuracy, while a lower learning rate improved convergence. The optimal learning rate for our model was 0.001.

The COCO dataset contains many object categories and complex scenes, which made training challenging. Data augmentation such as flipping, scaling, and color adjustments helped improve model generalization and reduced overfitting.
